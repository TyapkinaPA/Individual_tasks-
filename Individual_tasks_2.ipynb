```python
class Product:
  def __init__(self, name, cena, ves):
    self.name = name
    self.cena = cena
    self.ves = ves

class Buy(Product):
  def __init__(self, name, cena, ves, kol):
    super().__init__(name, cena, ves)
    self.kol = kol
    self.all_ves = kol * ves
    self.all_cena = kol* ves * cena

class Check(Buy):
  def info(self):
    print('Наименование товара: ', self.name)
    print('Цена товара: ', self.cena)
    print('Вес товара: ', self.ves)
    print('Количество товара: ', self.kol)
    print('Общий вес: ', self.all_ves)
    print('Общая стоимость: ', self.all_cena)
```

```python
products = [Buy('Яблоки', 100, 0.5, 10), Buy('Груши', 200, 0.25, 6), Buy('Апельсины', 150, 0.75, 3)]
for product in products:
  check = Check(product.name, product.cena, product.ves, product.kol)
  check.info()
  print()  
```

```python
class Person:
  def __init__(self, surname, name, patronymic):
    self.surname = surname
    self.name = name
    self.patronymic = patronymic
  def show_data(self):
    print('Фамилия: {}\n Имя: {}\n Отчество: {}'.format(self.surname, self.name, self.patronymic))
```

```python
class Student(Person):
  def __init__(self, surname, name, patronomic, average_score):
    super().__init__(surname, name, patronomic)
    self.average_score = average_score
  def show_data(self):
    super().show_data()
    print('Средний балл студента: {}'.format(self.average_score))
```

```python
class Professor(Person):
  def __init__(self, surname, name, patronomic, publications, post, age):
    super().__init__(surname, name, patronomic)
    self.publications = publications
    self.post = post
    self.age = age
  def show_data(self):
    super().show_data()
    print('Количество публикаций: {}'.format(self.publications))
    print('Должность: {}'.format(self.post))
    print('Возраст: {}'.format(self.age))  
```

```python
person = Person('Босенко', 'Тимур', 'Муртазович')
person.show_data()
print('-' * 7)
student = Student('Тяпкина', 'Полина', 'Андреевна', '1.9')
student.show_data()
print('-' * 7)
professor = Professor('Босенко', 'Тимур', 'Муртазович', '10', 'преподаватель, старший преподаватель, доцент, профессор', '38')
professor.show_data()
```
