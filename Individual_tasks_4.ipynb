```python
import datetime

def my_decorator(inner):
  def inner_decorator():
    print(datetime.datetime.utcnow())
        
    print(datetime.datetime.utcnow())

  return inner_decorator

@my_decorator
def recursive_sum(*args):
  if not args:
    return 0
  else:
    return args[0] + recursive_sum(*args[1:])
    
if __name__ == "__main__":
    recursive_sum()
```

```python
import datetime

def my_decorator(inner):
  def inner_decorator():
    print(datetime.datetime.utcnow())
        
    print(datetime.datetime.utcnow())

  return inner_decorator

@my_decorator
def answer(func):
  def kod(*args, **kwargs):
    result = func(*args, **kwargs)
    return "Результат функции: {}".format(result)
  return kod

if __name__ == "__main__":
    answer()
```

```python
@answer
def a_plus_b(a, b):
  return a + b
print(a_plus_b(3, 5))
print(a_plus_b(7, 9))
# Результат функции: 8
# Результат функции: 16
```

```python
@answer
def get_letters(text: str) -> str:
  return ''.join(sorted(set(filter(str.isalpha, text.lower()))))
print(get_letters('Hello, world!'))
print(get_letters('Декораторы это круто =)'))
# Результат функции: dehlorw
# Результат функции: адекортуыэ
```

```python
import datetime

def my_decorator(inner):
  def inner_decorator():
    print(datetime.datetime.utcnow())
    print(datetime.datetime.utcnow())

  return inner_decorator

@my_decorator
def repeater(func):
  def kod(*args, **kwargs):
    func(*args, **kwargs)
    func(*args, **kwargs)
  return kod

if __name__ == "__main__":
    repeater()
```

```python
@repeater
def multiply(num1, num2):
  print(num1 * num2)

multiply(2, 7) # после этого распечатается две строки со значением 14
multiply(5, 3) # после этого распечатается две строки со значением 15
```

```python
import resource
  
def limit_memory(maxsize):
  soft, hard = resource.getrlimit(resource.RLIMIT_AS)
  resource.setrlimit(resource.RLIMIT_AS, (maxsize, hard))

import time
start_time = time.time()

if __name__ == '__main__':
  limit_memory(64 * 1024 * 1024)
  def repeater(func):
    def kod(*args, **kwargs):
      func(*args, **kwargs)
      func(*args, **kwargs)
    return kod

  @repeater
  def multiply(num1, num2):
    print(num1 * num2)

  multiply(2, 7) # после этого распечатается две строки со значением 14
  multiply(5, 3) # после этого распечатается две строки со значением 15

end_time = time.time()
a = end_time - start_time 
print('\n Объем врмени', round(a, 5), 'cek')
import sys
a = [0] * (10 ** 6)
b = (sys.getsizeof(a)/(1024**2))
print('\n Объем памяти: ',round(b, 1), 'Mb')
```

```python
import datetime

def my_decorator(inner):
  def inner_decorator():
    print(datetime.datetime.utcnow())
        
    print(datetime.datetime.utcnow())

  return inner_decorator

@my_decorator
def validate_args(func):
    def kod(arg1, arg2):
      if not isinstance(arg1, int) or not isinstance(arg2, int):
        return "Wrong types"
      elif len([arg1, arg2]) < 2:
        return "Not enough arguments"
      elif len([arg1, arg2]) > 2:
        return "Too many arguments"
      else:
        return func(arg1, arg2)
    kod.__name__ = func.__name__
    kod.__doc__ = func.__doc__
    return kod

if __name__ == "__main__":
    validate_args()
```

```python
@validate_args
def my_function(a, b):
    """
    This function adds two integers
    """
    return a + b
```
